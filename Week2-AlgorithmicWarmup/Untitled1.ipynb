```python
import pandas as pd
```

```python
def fibonacci_number(n):
    assert 0 <= n <= 65
    if n <= 1:
        return n
    f_list = list(range(0, n+1))
    f_list[0] = 0
    f_list[1] = 1
    for i in range(2, n+1):
        f_list[i] = f_list[i-1] + f_list[i-2]
    return f_list[n]
```

```python
def last_digit_of_fibonacci_number(n):
    assert 0 <= n <= 10 ** 7
    if n <= 1:
        return n
    f_list = list(range(0, n+1))
    f_list[0] = 0
    f_list[1] = 1

    for i in range(2, n+1):
        f_list[i] = f_list[i-1] % 10 + f_list[i-2] % 10
    return f_list[n] % 10
```

```python
def last_digit_of_the_sum_of_fibonacci_numbers_naive(n):
    assert 0 <= n <= 10 ** 18

    if n <= 1:
        return n

    fibonacci_numbers = [0] * (n + 1)
    fibonacci_numbers[0] = 0
    fibonacci_numbers[1] = 1
    for i in range(2, n + 1):
        fibonacci_numbers[i] = fibonacci_numbers[i - 2] + fibonacci_numbers[i - 1]

    return sum(fibonacci_numbers) % 10
```

```python
def last_digit_of_the_sum_of_fibonacci_numbers_naive(n):
    assert 0 <= n <= 10 ** 18

    if n <= 1:
        return n

    fibonacci_numbers = [0] * (n + 1)
    fibonacci_numbers[0] = 0
    fibonacci_numbers[1] = 1
    for i in range(2, n + 1):
        fibonacci_numbers[i] = fibonacci_numbers[i - 2] + fibonacci_numbers[i - 1]

    return sum(fibonacci_numbers) % 10
```

```python
def sum_of_fibonacci_numbers_naive(n):
    if n <= 1:
        return n

    fibonacci_numbers = [0] * (n + 1)
    fibonacci_numbers[0] = 0
    fibonacci_numbers[1] = 1
    for i in range(2, n + 1):
        fibonacci_numbers[i] = fibonacci_numbers[i - 2] + fibonacci_numbers[i - 1]  
    return sum(fibonacci_numbers)
```

```python
def p_period_length(m):
    previous, current = 0, 1
    for i in range(0, m**2):
    for i in range(0, m**2):
        previous, current = current, (previous + current) % m
        if previous == 0 and current == 1:
            return i + 1
```

```python
def fibonacci_number_again(n, m):
    assert 0 <= n <= 10 ** 18 and 2 <= m <= 10 ** 3
    p_period = p_period_length(m)
    n = n % p_period
    if n <= 1:
        return n
    previous, current = 0, 1
    for _ in range(n - 1):
        previous, current = current, (previous + current)
    return current % m
```

```python
def generate_fibonacci(n):
    fn = []
    lasts = []
    last_sums = []
    sums = []
    last_sums_opt = []
    last_sums_new = []
    mods = []
    for i in range (0,n+1):
        fn.append(fibonacci_number(i))
        lasts.append(last_digit_of_fibonacci_number(i))
        last_sums.append(last_digit_of_the_sum_of_fibonacci_numbers_naive(i))
        sums.append(sum_of_fibonacci_numbers_naive(i))
        last_sums_opt.append(last_digit_of_the_sum_of_fibonacci_numbers(i))
        last_sums_new.append(last_digit_of_the_sum_of_fibonacci_numbers_new(i,10) )
        mods.append(fibonacci_number_again(i,60))
    return fn, lasts, last_sums, sums, last_sums_opt, last_sums_new, mods
```

```python
numbers, lasts, last_sums, sums, last_sums_opt, last_sums_new, mods = generate_fibonacci(65)
```
